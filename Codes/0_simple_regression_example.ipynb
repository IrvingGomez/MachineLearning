```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dat = pd.read_csv('Howell1.csv', sep=';')
dat = dat[dat['age']>=18]
```

```python
plt.figure(figsize=(10,7.5))
sns.scatterplot(data=dat, x='weight', y='height')
```

```python
bar_dat = dat.mean()

num = ((dat['weight']-bar_dat['weight'])*((dat['height']-bar_dat['height']))).sum()
den = ((dat['weight']-bar_dat['weight'])**2).sum()

hat_beta1 = num/den
hat_beta0 = bar_dat['height']-hat_beta1*bar_dat['weight']

weight_min = dat['weight'].min()
weight_max = dat['weight'].max()
```

```python
plt.figure(figsize=(10,7.5))
sns.scatterplot(data=dat, x='weight', y='height')
plt.plot([weight_min, weight_max],
    [hat_beta0 + hat_beta1*weight_min, hat_beta0 + hat_beta1*weight_max],
    linestyle='--', color='r', linewidth=4
)
#plt.savefig('weight_height_regression.png')
```
