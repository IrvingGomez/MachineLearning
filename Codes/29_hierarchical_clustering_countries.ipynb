```python
import pandas as pd
import matplotlib.pyplot as plt

import scipy.cluster.hierarchy as shc
from sklearn.preprocessing import normalize

data = pd.read_csv('Data_164_countries.csv')
data.head()
```

```python
country_names = data['Country']
country_names = country_names.tolist()

data = data.drop('Country', axis=1)

data_scaled = normalize(data)
data_scaled = pd.DataFrame(data_scaled, columns=data.columns)
data_scaled.head()
```

```python
plt.figure(figsize=(30, 20))
plt.title("Dendrogram, using complete")
dend = shc.dendrogram(shc.linkage(data_scaled, method='complete'), labels=country_names)
plt.xticks(fontsize=12)
```

```python
plt.figure(figsize=(30, 20))
plt.title("Dendrogram, using average")
dend = shc.dendrogram(shc.linkage(data_scaled, method='average'), labels=country_names)
plt.xticks(fontsize=12)
```

```python
plt.figure(figsize=(30, 20))
plt.title("Dendrogram, using single")
dend = shc.dendrogram(shc.linkage(data_scaled, method='single'), labels=country_names)
plt.xticks(fontsize=12)
```

```python
plt.figure(figsize=(30, 20))
plt.title("Dendrograms")
dend = shc.dendrogram(shc.linkage(data_scaled, method='ward'), labels=country_names)
plt.xticks(fontsize=12)
```
