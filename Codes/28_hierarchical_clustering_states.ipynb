```python
import pandas as pd
import matplotlib.pyplot as plt

import scipy.cluster.hierarchy as shc
```

```python
data = pd.read_csv('dat_simplify_grp_w_states.csv')
data.head()
```

```python
state_names = data['ent']
state_names = state_names.tolist()

data = data.drop('ent', axis=1)
```

```python
plt.figure(figsize=(30, 20))
plt.title("Dendrogram, using complete")
dend = shc.dendrogram(shc.linkage(data, method='complete'), labels=state_names)
plt.xticks(fontsize=12)
```

```python
plt.figure(figsize=(30, 20))
plt.title("Dendrogram, using average")
dend = shc.dendrogram(shc.linkage(data, method='average'), labels=state_names)
plt.xticks(fontsize=12)
```

```python
plt.figure(figsize=(30, 20))
plt.title("Dendrogram, using single")
dend = shc.dendrogram(shc.linkage(data, method='single'), labels=state_names)
plt.xticks(fontsize=12)
```

```python
plt.figure(figsize=(30, 20))
plt.title("Dendrograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'), labels=state_names)
plt.xticks(fontsize=12)
```
